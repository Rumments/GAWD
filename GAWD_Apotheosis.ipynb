# GAWD / GawdApotheosis — VARIABLE KEY

**formula**

Chemical formula used for the computation.


**Ex_pred_MeV**

Ex (MeV). Predicted pair–Compton crossover energy: the photon energy Ex where the model places pair production and Compton (incoherent) scattering at equal contribution for this composition.


**EPC_pred_MeV**

EPC (MeV). Predicted photoelectric crossover energy from the core closed-form mapping: the photon energy EPC where the model places the photoelectric term equal to the competing (non-photoelectric) term used in the EPC definition.


**EPC_shell2_MeV**

EPC (MeV) from the composite EPC mapping. This value equals EPC_pred_MeV unless a regime-specific EPC component is activated. When activated, the composite mapping uses one of two branches (low_gate or high_gate) whose magnitude is driven by Dlow or Dhigh.


**EPC_final_MeV**

EPC (MeV) after all enabled EPC components are applied. In most cases EPC_final_MeV equals EPC_shell2_MeV; it differs only when an additional domain-limited EPC component is enabled (e.g., edge-proximity handling for extreme-Z regimes).


**low_gate**

Indicator that the composite EPC mapping activated its low-branch EPC component for this composition.


**high_gate**

Indicator that the composite EPC mapping activated its high-branch EPC component for this composition.


**Dlow**

Dimensionless driver magnitude for the low-branch EPC component. Larger Dlow means stronger contribution from the low-branch component in EPC_shell2_MeV.


**Dhigh**

Dimensionless driver magnitude for the high-branch EPC component. Larger Dhigh means stronger contribution from the high-branch component in EPC_shell2_MeV.


**max_Z**

Maximum atomic number present in the formula.


**max_nsub**

Maximum outer-shell electron count among elements in the formula (taken from the periodic-table shells list). Used as a regime/complexity indicator inside the EPC composite mapping.


**Sclosest_pred**

Dimensionless edge-proximity score: a measure of how close the composition lies (in the model’s edge-proximity calculation) to relevant absorption-edge energies. Used only by domain-limited edge handling when enabled.


**w_mass_fractions**

Per-element mass fractions used by the model. Values sum to ~1.0. Keys are element symbols as stored by the pipeline (often lowercase).


**J1, J2, J3, J4**

Mass-fraction moments used as the core descriptor basis. For p in {1,2,3,4}:
Jp = sum over elements i of (w_i * (Z_i^p / A_i)),
where w_i is the element mass fraction, Z_i is atomic number, and A_i is atomic mass.


**Variance_surrogate**

Moment-derived variance proxy:
Variance_surrogate = J2 - (J1^2).


**R1S_array**

Per-element list of a 1s-scale proxy used internally by the EPC composite mapping (one value per element present, in the model’s internal order).


**R1s_weighted**

Single scalar derived from R1S_array by mass-fraction weighting; used as an internal driver in the EPC composite mapping.


**Shift_exponent**

Internal exponent parameter for the shift/tweak path (0.0 means no shift-exponent effect in the current configuration).


**VectorV3**

Convenience vector (J1, J2, J4) when enabled.


**VectorV4**

Convenience vector (J1, J2, J3, J4) when enabled.


**Etot**

Derived total energy coordinate (MeV):
Etot = Ex_pred_MeV + EPC_final_MeV.


REB
Derived relative energy balance (unitless):
REB = EPC_final_MeV / Etot.


Sigma
Derived unitless scalar computed deterministically from GAWD outputs to summarize response “shape/spread” behavior for downstream grouping and filtering (Apotheosis convenience metric).


delta_AN_raw
Derived unitless scalar computed deterministically from GAWD outputs to quantify the actinide/surrogate-shift signal used in your surrogate searches (Apotheosis convenience metric).


**Gain**

Dimensionless multiplicative factor produced by the EPC composite mapping logic (1.0 means no gain effect).


**shell_val**

Discrete regime label (integer) used to group results by the EPC composite mapping regime path selected (Apotheosis convenience label).
